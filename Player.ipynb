```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("nba2k-full.csv");df
```

```python
df.head(85)
```

```python
df = df.dropna();df
```

```python
#Drop Column

df = df.drop(["full_name","b_day","height","weight","college"], axis = 1)
df.head()
```

```python
#Creating function for removing "#","$" from jersey and salary.
def removehash(value):
    value = value[1:]
    return int(value)

df['jersey'] = df['jersey'].apply(removehash)
df['salary'] = df['salary'].apply(removehash)
df.head()
```

```python
df['country'].value_counts()
```

```python
def removecountryoutlier(value):
    if value not in ['USA', 'Canada', 'Australia']:
        return 'Others'
    else:
        return value

df['country'] = df['country'].apply(removecountryoutlier)
df['country'].value_counts()
```

```python
def removeundrafted(value):
    if value=='Undrafted':
        return 
    else:
        return value
    
df['draft_round'] = df['draft_round'].apply(removeundrafted)
df['draft_peak'] = df['draft_peak'].apply(removeundrafted)

df = df.dropna()

def removehash(value):
    value = value[1:]
    return int(value)

#Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['position'] = le.fit_transform(df['position'])
df['country'] = le.fit_transform(df['country'])
df['team'] = le.fit_transform(df['team'])

df.head()
```

```python
x = df.drop(['salary'], axis = 1)
y = df['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regr_1 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(x, y)
# Predict
y_1 = regr_1.predict(x_test)
r2_score(y_test,y_1)
```
